```python
import json
import os
import sys
```

```python
# Load credentials from json file
with open("twitter_credentials.json", "r") as file:  
    creds = json.load(file)

client_key = creds['CONSUMER_KEY']
client_secret = creds['CONSUMER_SECRET']
```

```python
import base64

key_secret = '{}:{}'.format(client_key, client_secret).encode('ascii')
b64_encoded_key = base64.b64encode(key_secret)
b64_encoded_key = b64_encoded_key.decode('ascii')
```

```python
import requests

base_url = 'https://api.twitter.com/'
auth_url = '{}oauth2/token'.format(base_url)

auth_headers = {
    'Authorization': 'Basic {}'.format(b64_encoded_key),
    'Content-Type': 'application/x-www-form-urlencoded;charset=UTF-8'
}

auth_data = {
    'grant_type': 'client_credentials'
}

auth_resp = requests.post(auth_url, headers=auth_headers, data=auth_data)

# Check status code okay
auth_resp.status_code
```

```python
# Keys in data response are token_type (bearer) and access_token (your access token)
auth_resp.json().keys()
```

```python
access_token = auth_resp.json()['access_token']

search_headers = {
    'Authorization': 'Bearer {}'.format(access_token)    
}

search_params = {'query': '@PizzaToThePolls',
                'fromDate': '201811060000',
                'toDate': '201811062359',
                'maxResults': 500}
count_params = search_params.update({'bucket': 'hour'})

search_url = '{}1.1/tweets/search/fullarchive/research.json'.format(base_url)
count_url = '{}1.1/tweets/search/fullarchive/research/counts.json'.format(base_url)


count_resp = requests.get(count_url, headers = search_headers, params = search_params)
count_data = count_resp.json()
```
